# Sales Prediction using Multiple regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the dataset

```python
ds = pd.read_csv('Advertising.csv')
```

# Checking for Outliers

```python
fig, axs = plt.subplots(3, figsize = (5,5))
plt1 = sns.boxplot(ds['TV'], ax = axs[0])
plt2 = sns.boxplot(ds['newspaper'], ax = axs[1])
plt3 = sns.boxplot(ds['radio'], ax = axs[2])
plt.tight_layout()
```

# Checking how individual feature affects the sales

```python
sns.pairplot(ds, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', height=4, aspect=1, kind='scatter')
plt.show()
```

# Heatmap

```python
sns.heatmap(ds.corr(),annot = True)
```

# Setting the value of x and y

```python
x = ds[['TV', 'radio', 'newspaper']]      # independent variable
y = ds['sales']                           # dependent variable
```

# Data Splitting 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=0) 
```

# Model Building

```python
mlr = LinearRegression()
mlr.fit(x_train,y_train)

y_pred = mlr.predict(x_test)
```

# Performance of the Model

```python
from sklearn import metrics

MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMS = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

print('Mean Absolute Error:', MAE)
print('Mean Square Error:', MSE)
print('Root Mean Square Error:',RMS)
```
